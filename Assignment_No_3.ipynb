```python
# • Find common elements between two arrays [with and without using intersect1d, use isin() instead]

arr1 = np.array([0, 10, 20, 40, 60, 10])
arr2 = np.array([10, 30, 40, 50])
arr1
```

```python
arr2
```

```python
#Intersected

common = np.intersect1d(arr1,arr2)
common
```
